```python
def MI(y1, y2):
    iterations = 25
    nknns = 20
    sample = 2000
    Mi = np.zeros(iterations)
    for i in range(iterations):
        index_sample = np.random.randint(0, len(y1), sample) 
        Mi_i = mutual_info_regression(y1[index_sample].reshape(-1, 1) , y2[index_sample] , n_neighbors=nknns, copy=True, random_state=None, n_jobs=-1)
        Mi[i] = Mi_i[0]/np.log(2)

    return Mi.mean(), Mi.std()

def RMSE_sample(y1,y2):
    iterations = 25
    sample = 2000
    RMSEi = np.zeros(iterations)
    for i in range(iterations):
        index_sample = np.random.randint(0, len(y1), sample) 
        RMSE_i = ((y1[index_sample] - y2[index_sample])**2).mean()**0.5
        RMSEi[i] = RMSE_i

    return RMSEi.mean(), RMSEi.std()
```

```python
from matplotlib.colors import ListedColormap, BoundaryNorm
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Define the colors and corresponding ranges
colors = ['white', 'lavender', 'lightblue', 'lightskyblue', 'dodgerblue', 'royalblue', 'blue', 'darkblue']
ranges = [0, 2, 25, 50, 200, 500, 750, 1000, 1300]  # Define 8 intervals

# Create the colormap using ListedColormap
cmap = ListedColormap(colors)

# Create BoundaryNorm to map values to the colormap based on the defined ranges
norm = BoundaryNorm(boundaries=ranges, ncolors=len(colors))

fig = plt.figure(figsize=(15, 3))
gs = GridSpec(1, 4, figure=fig, width_ratios=[1, 1, 1, 0.04])  # Add space for the colorbar

vmax = 1300
bins = 500

# First subplot
ax1 = fig.add_subplot(gs[0, 0]) 
img1 = ax1.imshow(obs_test.swe.max(dim='time'), origin='lower', cmap=cmap, norm=norm)
ax1.set_title('a) Training only with SNOTEL')

# Second subplot
ax2 = fig.add_subplot(gs[0, 1]) 
img2 = ax2.imshow((true_test.swe * mask).max(dim='time'), origin='lower', cmap=cmap, norm=norm)
ax2.set_title('b) Parflow-CLM true')
ax2.set_yticks([])

# Third subplot
ax3 = fig.add_subplot(gs[0, 2]) 
img3 = ax3.imshow(fine_test.swe.max(dim='time'), origin='lower', cmap=cmap, norm=norm)
ax3.set_title('c) Finetuned from EXP1 to SNOTEL')
ax3.set_yticks([])

# Add a single colorbar for all subplots
cbar_ax = fig.add_subplot(gs[0, 3])  # Extra axis for colorbar
cbar = fig.colorbar(img1, cax=cbar_ax, boundaries=ranges)
cbar.ax.set_ylabel('SWE (mm)')  # Add label to colorbar

plt.tight_layout()  # Adjust layout to avoid overlap

plt.savefig('Fig4.jpg', bbox_inches='tight', pad_inches=0.1)
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes


HUC2 = rioxarray.open_rasterio('/hydrodata/national_mapping/CONUS1/HUC2_CONUS1_grid.tif')
plt.figure(figsize=(20,30))
ii=1
# Main plot with imshow
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(14,6))
for i in [14,7]:
    mask = HUC2.isel(x=slice(0,3200), y=slice(0,1800)) == i
    mask = xr.concat([mask] * 360, dim='time').squeeze('band')
    mask = mask.where(mask, 0)
    if mask.sum() != 0:
        mask['time'] = true_test.swe.time
        mask = mask.drop_vars('band')
        mask['x'] = true_test.swe.x
        mask['y'] = true_test.swe.y

        clipped_parflow = true_test.swe * mask
        clipped_y_hat = obs_test.swe * mask
        clipped_y_hat2 = fine_test.swe * mask

        scale_factor = (1/mask.isel(time=0).sum()).values

        plt.subplot(1,2,ii)
        (clipped_parflow.sum(dim={'x','y'}) * scale_factor).plot(label='ParFlow-CLM true')
        (clipped_y_hat.sum(dim={'x','y'}) * scale_factor).plot(label='Training only with\nSNOTEL')
        (clipped_y_hat2.sum(dim={'x','y'}) * scale_factor).plot(label='Finetuned from EXP1\nto SNOTEL')
        max_swe = (clipped_parflow.sum(dim={'x','y'}) * scale_factor).values.max()
        if ii != 1:
            plt.legend()
        plt.xlabel('time')
        plt.ylabel('SWE (mm)')
        plt.title('HUC2 - ' + HUC2_names[str(i)])
        plt.grid()

        # Create inset axes
        inset_ax = inset_axes(ax[ii-1], width="25%", height="15%", loc='upper left', borderpad=2)

        # Add an image to the inset plot
        inset_data = np.random.rand(5, 5)
        inset_image = inset_ax.imshow(mask.isel(time=0), cmap='Blues')

        # Optional: Adjust aspect ratio if needed
        inset_ax.set_aspect('auto')  # or 'equal', depending on your needs

        # Optional: Remove ticks and labels from inset
        inset_ax.set_xticks([])
        inset_ax.set_yticks([])
        
        inset_ax.set_title('Basin', fontsize=10)
        
        ii = ii + 1

plt.savefig('Fig5.jpg', bbox_inches='tight', pad_inches=0.1)
```
