```python
def MI(y1, y2):
    iterations = 25
    nknns = 20
    sample = 2000
    Mi = np.zeros(iterations)
    for i in range(iterations):
        index_sample = np.random.randint(0, len(y1), sample) 
        Mi_i = mutual_info_regression(y1[index_sample].reshape(-1, 1) , y2[index_sample] , n_neighbors=nknns, copy=True, random_state=None, n_jobs=-1)
        Mi[i] = Mi_i[0]/np.log(2)

    return Mi.mean(), Mi.std()

def RMSE_sample(y1,y2):
    iterations = 25
    sample = 2000
    RMSEi = np.zeros(iterations)
    for i in range(iterations):
        index_sample = np.random.randint(0, len(y1), sample) 
        RMSE_i = ((y1[index_sample] - y2[index_sample])**2).mean()**0.5
        RMSEi[i] = RMSE_i

    return RMSEi.mean(), RMSEi.std()
```
